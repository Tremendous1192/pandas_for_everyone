## 3.4 seaborn

### データを読み込む

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context("paper")# 印刷用に最適化する(デフォルトはnotebook)

# データを読み込む
tips = sns.load_dataset("tips")
tips
```

### ヒストグラム

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.histplot(data = tips, x = "total_bill", ax = ax)

# タイトル等
ax.set_title("Total Bill Histgram")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Count")

plt.show()
```

### KDEプロット(密度プロット)

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.kdeplot(data = tips, x = "total_bill", ax = ax)

# タイトル等
ax.set_title("Total Bill Density")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Unit Probability")

plt.show()
```

### ラグプロットとヒストグラム

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.rugplot(data = tips, x = "total_bill", ax = ax)
sns.histplot(data = tips, x = "total_bill", ax = ax)

# タイトル等
ax.set_title("Total Bill Histgram")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Count")

plt.show()
```

### 分布プロット

```python
# グラフを描画する
fig = sns.displot(data = tips, x = "total_bill", kde = True, rug = True)

# タイトル等
fig.set_axis_labels(x_var = "Total Bill", y_var = "Count")
fig.figure.suptitle("Distribution of Total Bill")

plt.show()
```

### 棒グラフ

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.countplot(data = tips, x = "day", legend = False, ax = ax)

# タイトル等
ax.set_title("Count of days")
ax.set_xlabel("Day of the Week")
ax.set_ylabel("Frequency")

plt.show()
```

### 散布図

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.scatterplot(data = tips, x = "total_bill", y = "tip", ax = ax)

# タイトル等
ax.set_title("Scatter Plot of Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

### Regression Plot

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.regplot(data = tips, x = "total_bill", y = "tip", ax = ax)

# タイトル等
ax.set_title("Regression Plot of Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

### 連結プロット

```python
# グラフを描画する
fig = sns.jointplot(data = tips, x = "total_bill", y = "tip")

# タイトル等
fig.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
fig.figure.suptitle("Joint Plot of Total Bill vs Tip", y = 1.03)

plt.show()
```

### hexbinプロット(六角形ビニング)

```python
# グラフを描画する
fig = sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "hex")

# タイトル等
fig.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
fig.figure.suptitle("Hexbin Plot of Total Bill vs Tip", y = 1.03)

plt.show()
```

### 2次元の密度プロット

#### kdeplotを用いた密度プロット

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.kdeplot(data = tips, x = "total_bill", y = "tip", fill = True, ax = ax)

# タイトル等
ax.set_title("Kernel Density Plot of Total Bill vs Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")

plt.show()
```

### 棒グラフ

```python
import numpy as np

#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.barplot(data = tips, x = "time", y = "total_bill", estimator = np.mean, ax = ax)

# タイトル等
ax.set_title("Bar Plot of Total Boll for Time of Day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Average Total Bill")

plt.show()
```

### 箱ひげ図

### バイオリンプロット

```python
#キャンバスとサブプロット
fig, ax = plt.subplots()

# グラフの描画
sns.violinplot(data = tips, x = "time", y = "total_bill", ax = ax)

# タイトル等
ax.set_title("Box Plot of Total Boll for Time of Day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Average Total Boll")

plt.show()
```

### 箱ひげ図とバイオリンプロットを並べる

```python
#キャンバスとサブプロット
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)

# グラフの描画
sns.boxplot(data = tips, x = "time", y = "total_bill", ax = ax1)
sns.violinplot(data = tips, x = "time", y = "total_bill", ax = ax2)

# タイトル等
ax1.set_title("Box Plot of Total Boll for Time of Day")
ax1.set_xlabel("Time of Day")
ax1.set_ylabel("Average Total Boll")

ax2.set_title("Box Plot of Total Boll for Time of Day")
ax2.set_xlabel("Time of Day")
ax2.set_ylabel("Average Total Boll")

fig.suptitle("Comparison of Box Plot with Violin Plot")

plt.show()
```

### ペアの相関を見る

```python
# グラフを描画する
fig = sns.pairplot(data = tips)

# タイトル等
fig.figure.suptitle("Pairwise Relationships of the Tips Data", y = 1.03)

plt.show()
```

### PairGridを用いてペアの相関を見る

```python
# グラフを描画する
fig = sns.PairGrid(data = tips, diag_sharey = False)

# 上下あるいは斜め隣の描画
fig = fig.map_upper(sns.regplot)
fig = fig.map_diag(sns.histplot)
fig = fig.map_lower(sns.kdeplot)

# タイトル等
fig.figure.suptitle("Pairwise Relationships of the Tips Data", y = 1.03)

plt.show()
```

### 2個のファセット変数を用いた散布図

```python
# 性別に色を割り当てる
colors = {
    "Yes": "#f1a340",# オレンジ
    "No": "#998ec3",# 紫
}

# グラフを描画する
fig = sns.relplot(data = tips,
                  x = "total_bill",
                  y = "tip",
                  hue = "smoker",
                  style = "sex",
                  kind = "scatter",
                  col = "day",
                  row = "time",
                  palette = colors,
                  aspect = 1.7
                  )

# タイトル等
fig.set_titles(
    row_template = "{row_name}",
    col_template = "{col_name}"
)
fig.figure.suptitle("Scatter Plot of the Tips Data", y = 1.03)

# レイアウト調整
sns.move_legend(
    fig,
    loc = "lower center",
    bbox_to_anchor = (0.5, 1),
    ncol = 2,
    title = None,
    frameon = False
)
fig.figure.set_tight_layout(True)

plt.show()
```

### スタイルの違いをバイオリンプロットで確認する

```python
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# キャンバス
fig = plt.figure()

# スタイル別のグラフを表示する
for idx,style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(
            data = tips, x = "time", y = "total_bill", ax = ax
        )
        violin.set_title(style)

# レイアウト調整
fig.set_tight_layout(True)

plt.show()
```

### コンテキストの違いをバイオリンプロットで確認する

```python
import pandas as pd
contexts = pd.DataFrame({
    "paper": sns.plotting_context("paper"),
    "notebook": sns.plotting_context("notebook"),
    "talk": sns.plotting_context("talk"),
    "poster": sns.plotting_context("poster")
})

context_styles = contexts.columns

# キャンバス
fig = plt.figure()

# スタイル別のグラフを表示する
for idx,style in enumerate(context_styles):
    plot_position = idx + 1
    with sns.plotting_context(style):
        ax = fig.add_subplot(2, 2, plot_position)
        violin = sns.violinplot(
            data = tips, x = "time", y = "total_bill", ax = ax
        )
        violin.set_title(style)

# レイアウト調整
fig.set_tight_layout(True)

plt.show()
```
